```python
import os
from slugify import slugify
import shutil
```

```python
book_dir = "/home/panagoa/book"
airflow_pdf_dir = "../data/pdfs"
```

```python
for orig_file_name in os.listdir(book_dir):
    if not orig_file_name.endswith("pdf"):
        continue

    new_file_name = slugify(orig_file_name.replace(".pdf", "")) + ".pdf"
    print(orig_file_name, new_file_name)
    shutil.copy(os.path.join(book_dir, orig_file_name), os.path.join(airflow_pdf_dir, new_file_name))
```
