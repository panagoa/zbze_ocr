```python

import os

pdf_path = '../data/dag_results/pdf_processing/dysche_zhyg.pdf'

all_words = set()
for lang_dir in os.listdir(pdf_path):
    if not lang_dir.startswith('rslt'):
        continue
        
    with open(os.path.join(pdf_path, lang_dir, 'words_rslt_not_exist_dictionary.txt'), 'r') as f:
        words = f.read().split('\n')
        words = set(words)
        all_words = all_words.union(words)
```

```python
with open('../data/dag_results/pdf_processing/all_words.txt', 'w') as f:
    f.write('\n'.join(all_words))
```
