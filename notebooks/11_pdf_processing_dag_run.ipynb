```python
import os
```

```python
airflow_pdf_dir = "../data/pdfs"
```

```python
from pprint import pprint
from airflow_client import client
from airflow_client.client.api import dag_run_api
from airflow_client.client.model.dag_run import DAGRun

configuration = client.Configuration(
    host="http://localhost:8080/api/v1", username="zbze", password=os.getenv("ZBZE_AIRFLOW_USER_PSWD")
)


with client.ApiClient(configuration) as api_client:
    api_instance = dag_run_api.DAGRunApi(api_client)

    for pdf_name in sorted(os.listdir(airflow_pdf_dir)):
        dag_id = "pdf_processing"
        dag_run = DAGRun(
            conf={
                "TESSERACT_LANG": "kbd",
                "TESSERACT_LANG_COMPARE": "kbd_ng",
                "PDF_NAME": pdf_name,
            },
            note="note_example",
        )

        try:
            api_response = api_instance.post_dag_run(dag_id, dag_run)
            pprint(api_response)
        except client.ApiException as e:
            print("Exception when calling DAGRunApi->post_dag_run: %s\n" % e)
```
