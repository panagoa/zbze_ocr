```python
import ahocorasick

word2morph_dawg = ahocorasick.Automaton()
```

```python
import json

with open('../dags/src/stemmer/tests/test_data/morph2words.json', 'r') as f:
    morph2words = json.load(f)
```

```python
from collections import defaultdict

word2morph = defaultdict(list)
for data in morph2words:
    for word in data['array_agg']:
        word2morph[word].append(data['morph'])

for word, morphs in word2morph.items():
    word2morph_dawg.add_word(word, morphs)
```

```python
word2morph_dawg.make_automaton()
```
